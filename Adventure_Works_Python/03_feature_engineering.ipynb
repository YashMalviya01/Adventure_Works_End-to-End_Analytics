# Date Feature Engineering

## Objective
The purpose of this section is to create date-based analytical features from the sales transaction data.

Date features are essential for:

Operational datasets typically store raw timestamps, but analytical workflows require structured date dimensions and time-based attributes.

# Goals of This Section

4. Extracting weekday information
5. Preparing datasets for time-series analysis

# Expected Outcome

At the end of this section:

```python
# Convert Date Columns

sales_order_header['orderdate'] = pd.to_datetime(
    sales_order_header['orderdate']
)

sales_order_header['duedate'] = pd.to_datetime(
    sales_order_header['duedate']
)

sales_order_header['shipdate'] = pd.to_datetime(
    sales_order_header['shipdate']
)

# Create Date Features

# Order Year
sales_order_header['order_year'] = (
    sales_order_header['orderdate'].dt.year
)

# Order Month Number
sales_order_header['order_month'] = (
    sales_order_header['orderdate'].dt.month
)

# Order Month Name
sales_order_header['order_month_name'] = (
    sales_order_header['orderdate'].dt.month_name()
)

# Order Quarter
sales_order_header['order_quarter'] = (
    sales_order_header['orderdate'].dt.quarter
)

# Order Day Name
sales_order_header['order_day_name'] = (
    sales_order_header['orderdate'].dt.day_name()
)

# Order Weekday Number
sales_order_header['order_weekday'] = (
    sales_order_header['orderdate'].dt.weekday
)

# Preview Engineered Features

sales_order_header[
    [
        'salesorderid',
        'orderdate',
        'order_year',
        'order_month',
        'order_month_name',
        'order_quarter',
        'order_day_name'
    ]
].head()
```

# Key Metrics

The following business metrics will be engineered:

| Total Cost | Estimated product cost |
| Profit | Revenue minus cost |
| Profit Margin | Profitability percentage |

# Expected Outcome

At the end of this section:

```python

# Merge Product Information

sales_fact = sales_order_detail.merge(
    product[
        [
            'productid',
            'standardcost',
            'listprice'
        ]
    ],
    on='productid',
    how='left'
)

# Revenue Metrics

# Revenue
sales_fact['revenue'] = (
    sales_fact['linetotal']
)

# Total Cost
sales_fact['total_cost'] = (
    sales_fact['standardcost'] *
    sales_fact['orderqty']
)

# Profit
sales_fact['profit'] = (
    sales_fact['revenue'] -
    sales_fact['total_cost']
)

# Profit Margin Percentage
sales_fact['profit_margin_pct'] = (
    (
        sales_fact['profit'] /
        sales_fact['revenue']
    ) * 100
).round(2)

# Average Unit Price
sales_fact['average_unit_price'] = (
    sales_fact['revenue'] /
    sales_fact['orderqty']
).round(2)

# Preview Engineered Metrics

sales_fact[
    [
        'salesorderid',
        'productid',
        'orderqty',
        'revenue',
        'total_cost',
        'profit',
        'profit_margin_pct',
        'average_unit_price'
    ]
].head()
```

# Product Hierarchy Engineering

## Objective
The purpose of this section is to enrich product-level transactional data with category and subcategory information.

Product hierarchy engineering helps transform isolated product records into business-friendly analytical dimensions.

This process enables:
- category-level analysis
- subcategory performance analysis
- product segmentation
- dashboard drilldowns
- executive reporting

---

# Goals of This Section

This section focuses on:

1. Merging product subcategory information
2. Merging product category information
3. Creating a complete product hierarchy
4. Enriching transactional sales data

---

# Business Importance

Businesses rarely analyze products individually at scale.

Instead, organizations commonly analyze:
- product categories
- product families
- subcategories
- product segments

This hierarchy enables:
- category-level KPIs
- product mix analysis
- sales contribution analysis
- inventory insights

---

# Expected Outcome

At the end of this section:
- products will contain category hierarchy information
- transactional datasets will become dimensionally enriched
- analytical reporting capabilities will improve significantly

```python
# Merge Product Subcategory

product_enriched = product.merge(
    product_subcategory[
        [
            'productsubcategoryid',
            'name',
            'productcategoryid'
        ]
    ],
    on='productsubcategoryid',
    how='left',
    suffixes=('', '_subcategory')
)

# Rename Subcategory Column
product_enriched.rename(
    columns={
        'name_subcategory': 'product_subcategory_name'
    },
    inplace=True
)

# Merge Product Category

product_enriched = product_enriched.merge(
    product_category[
        [
            'productcategoryid',
            'name'
        ]
    ],
    on='productcategoryid',
    how='left',
    suffixes=('', '_category')
)

# Rename Category Column
product_enriched.rename(
    columns={
        'name_category': 'product_category_name'
    },
    inplace=True
)

# Merge Enriched Product Data Into Sales Fact

sales_fact = sales_fact.merge(
    product_enriched[
        [
            'productid',
            'name',
            'product_subcategory_name',
            'product_category_name'
        ]
    ],
    on='productid',
    how='left'
)

# Rename Product Name Column
sales_fact.rename(
    columns={
        'name': 'product_name'
    },
    inplace=True
)

# Preview Product Hierarchy

sales_fact[
    [
        'productid',
        'product_name',
        'product_subcategory_name',
        'product_category_name',
        'revenue',
        'profit'
    ]
].head()
```
