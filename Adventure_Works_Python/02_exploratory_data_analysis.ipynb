# Why Missing Value Analysis Matters

In analytics projects, missing data can lead to:
- inaccurate revenue calculations
- incomplete customer insights
- broken dimensional relationships
- misleading visualizations

Performing null analysis before cleaning ensures:
- better data quality
- reliable business reporting
- stronger analytical models

---

# Expected Outcome

At the end of this section:
- important null-heavy columns will be identified
- datasets requiring cleaning will be recognized
- cleaning priorities will become clearer

```python
# Missing Value Analysis


for name, df in datasets.items():

    print("\n" + "=" * 70)
    print(f"MISSING VALUE ANALYSIS: {name.upper()}")
    print("=" * 70)

    # Total Missing Values
    missing_values = df.isnull().sum()

    # Missing Percentage
    missing_percentage = (
        (df.isnull().sum() / len(df)) * 100
    ).round(2)

    # Combine Results
    missing_summary = pd.DataFrame({
        'missing_values': missing_values,
        'missing_percentage': missing_percentage
    })

    # Filter Columns Having Missing Values
    missing_summary = missing_summary[
        missing_summary['missing_values'] > 0
    ].sort_values(
        by='missing_percentage',
        ascending=False
    )

    # Display Results
    if missing_summary.empty:
        print("\nNo missing values found.")
    else:
        display(missing_summary)
```

# Data Cleaning Strategy

The cleaning process follows a business-driven approach rather than blindly removing missing values.

Missing values will be handled using four strategies:

| Strategy | Purpose |
|---|---|
| Drop | Remove unusable columns |
| Keep | Preserve business-related null values |
| Impute | Fill missing values where appropriate |
| Flag | Convert missingness into business indicators |

---

# Important Cleaning Principles

This project treats missing values based on business meaning.

Examples:
- some null values represent valid operational behavior
- some columns are optional business fields
- some attributes apply only to specific product categories

This ensures that important business information is not lost during cleaning.

```python
# =========================================================
# Standardize Column Names
# =========================================================

for name, df in datasets.items():

    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(' ', '_')
    )

# =========================================================
# Drop Unnecessary Columns
# =========================================================

# sales_order_header
if 'comment' in sales_order_header.columns:

    sales_order_header.drop(
        columns=['comment'],
        inplace=True
    )

# product
drop_columns_product = [
    'discontinueddate'
]

existing_columns = [
    col for col in drop_columns_product
    if col in product.columns
]

product.drop(
    columns=existing_columns,
    inplace=True
)

# =========================================================
# Handle Missing Values
# =========================================================

# Product Table
if 'color' in product.columns:

    product['color'] = product['color'].fillna(
        'Unknown'
    )

if 'size' in product.columns:

    product['size'] = product['size'].fillna(
        'Not Specified'
    )

if 'style' in product.columns:

    product['style'] = product['style'].fillna(
        'Unknown'
    )

if 'class' in product.columns:

    product['class'] = product['class'].fillna(
        'Unknown'
    )

if 'productline' in product.columns:

    product['productline'] = product['productline'].fillna(
        'Unknown'
    )

if 'weight' in product.columns:

    product['weight'] = product['weight'].fillna(
        product['weight'].median()
    )

# =========================================================
# Create Business Flags
# =========================================================

# Online Orders
if 'salespersonid' in sales_order_header.columns:

    sales_order_header['is_online_order'] = (
        sales_order_header['salespersonid']
        .isnull()
        .astype(int)
    )

# Store Customers
if 'storeid' in customer.columns:

    customer['is_store_customer'] = (
        customer['storeid']
        .notnull()
        .astype(int)
    )

# =========================================================
# Convert Date Columns
# =========================================================

date_columns_sales = [
    'orderdate',
    'duedate',
    'shipdate'
]

for col in date_columns_sales:

    if col in sales_order_header.columns:

        sales_order_header[col] = pd.to_datetime(
            sales_order_header[col]
        )

# =========================================================
# Remove Duplicate Rows
# =========================================================

for name, df in datasets.items():

    before_rows = df.shape[0]

    df.drop_duplicates(inplace=True)

    after_rows = df.shape[0]

    print(
        f"{name}: Removed {before_rows - after_rows} duplicate rows"
    )

# =========================================================
# Verify Missing Values After Cleaning
# =========================================================

for name, df in datasets.items():

    print("\n" + "=" * 70)
    print(f"CLEANED DATASET: {name.upper()}")
    print("=" * 70)

    missing_values = df.isnull().sum()

    missing_values = missing_values[
        missing_values > 0
    ]

    if missing_values.empty:

        print("\nNo missing values remaining.")

    else:

        display(
            missing_values.sort_values(
                ascending=False
            )
        )

```

```python
# =========================================================
# Export Cleaned Datasets
# =========================================================

base_path = '/Users/yashmalviya/AdventureWorks-Docker/01_data/02_processed/'

sales_order_header.to_csv(
    base_path + 'sales_order_header_cleaned.csv',
    index=False
)

sales_order_detail.to_csv(
    base_path + 'sales_order_detail_cleaned.csv',
    index=False
)

customer.to_csv(
    base_path + 'customer_cleaned.csv',
    index=False
)

product.to_csv(
    base_path + 'product_cleaned.csv',
    index=False
)

product_category.to_csv(
    base_path + 'product_category_cleaned.csv',
    index=False
)

product_subcategory.to_csv(
    base_path + 'product_subcategory_cleaned.csv',
    index=False
)

sales_territory.to_csv(
    base_path + 'sales_territory_cleaned.csv',
    index=False
)


# =========================================================
# Cleaning Process Completed
# =========================================================

print("\n" + "=" * 70)
print("DATA CLEANING COMPLETED SUCCESSFULLY")
print("=" * 70)

print("\nCleaned datasets exported successfully to:")
print(base_path)
```
