The project primarily revolves around the sales analytics workflow.

---

# Important Table Granularity

| Table | Granularity |
|---|---|
| sales_order_header | One row per order |
| sales_order_detail | One row per product inside an order |
| customer | One row per customer |
| product | One row per product |
| product_subcategory | One row per product subcategory |
| product_category | One row per product category |
| sales_territory | One row per sales territory |

Understanding granularity is critical before:
- joins
- aggregations
- KPI calculations

---

# Key Relationships

| Parent Table | Child Table | Key |
|---|---|---|
| sales_order_header | sales_order_detail | SalesOrderID |
| customer | sales_order_header | CustomerID |
| product | sales_order_detail | ProductID |
| product_subcategory | product | ProductSubcategoryID |
| product_category | product_subcategory | ProductCategoryID |
| sales_territory | sales_order_header | TerritoryID |

---

# Why This Matters

Understanding relationships helps prevent:
- duplicate aggregations
- incorrect joins
- inflated revenue calculations
- inaccurate KPIs

This phase establishes the analytical foundation required for:
- exploratory analysis
- feature engineering
- dimensional modeling
- SQL analytics
- dashboard development

```python
# =========================================================
# Initial Dataset Understanding
# =========================================================

for name, df in datasets.items():

    print("\n" + "=" * 70)
    print(f"UNDERSTANDING DATASET: {name.upper()}")
    print("=" * 70)

    # Basic Information
    print(f"\nNumber of Rows: {df.shape[0]}")
    print(f"Number of Columns: {df.shape[1]}")

    # Memory Usage
    print("\nMemory Usage:")
    print(df.memory_usage(deep=True).sum() / 1024**2, "MB")

    # Duplicate Rows
    print("\nDuplicate Rows:")
    print(df.duplicated().sum())

    # Numerical Summary
    print("\nNumerical Summary:")
    display(df.describe())

    # Categorical Summary
    print("\nCategorical Summary:")
    display(df.describe(include='object'))
```
